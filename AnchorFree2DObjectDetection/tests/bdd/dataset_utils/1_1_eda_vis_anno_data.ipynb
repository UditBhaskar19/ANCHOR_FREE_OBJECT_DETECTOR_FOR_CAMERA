```python
import os, sys
module_rootdir = '../../..'
dataset_root_dir = '../../../../'
sys.path.append(module_rootdir)

import config_dataset
from modules.dataset_utils.bdd_dataset_utils.bdd_utils import aggregate_image_path, aggregate_ground_truths
from modules.plot.viz_annotation import vizualize_annotations
```

```python
train_images_dir = config_dataset.bdd_train_images_dir
val_images_dir = config_dataset.bdd_val_images_dir
test_images_dir = config_dataset.bdd_test_images_dir

dataset_meta_info = aggregate_image_path(
    dataset_root_dir, 
    train_images_dir, 
    val_images_dir, 
    test_images_dir)

dataset_meta_info
```

```python
train_labels_file = config_dataset.bdd_train_labels_file
train_lane_labels_file = config_dataset.bdd_train_lane_labels_file

val_labels_file = config_dataset.bdd_val_labels_file
val_lane_labels_file = config_dataset.bdd_val_lane_labels_file

selected_labels, label_names = aggregate_ground_truths(
    dataset_root_dir,
    train_labels_file, 
    train_images_dir, 
    train_lane_labels_file, 
    verbose=False)

selected_labels, label_names = aggregate_ground_truths(
    dataset_root_dir,
    val_labels_file, 
    val_images_dir, 
    val_lane_labels_file, 
    verbose=False)
```

```python
# for idx in range(50, 65):
for idx in range(0, 10):
    
    img_path = os.path.join(dataset_root_dir, selected_labels[idx]['img_path'])
    lane_img_path = os.path.join(dataset_root_dir, selected_labels[idx]['lane_img_path'])
    
    polygon_coord = selected_labels[idx]['drivableArea']
    bboxs = selected_labels[idx]['boundingBox2D']
    class_labels = selected_labels[idx]['objCategory']
    
    weather = selected_labels[idx]['weather']
    scene = selected_labels[idx]['scene']
    timeofday = selected_labels[idx]['timeofday']
    
    vizualize_annotations(
        img_path, bboxs, class_labels, 
        lane_img_path, polygon_coord,
        alpha = 0.9, beta = 1.0,
        figsize=(10, 5))
    
    print('weather   :', weather)
    print('scene     :', scene)
    print('timeofday :', timeofday)
    
    print('=' * 100)
```

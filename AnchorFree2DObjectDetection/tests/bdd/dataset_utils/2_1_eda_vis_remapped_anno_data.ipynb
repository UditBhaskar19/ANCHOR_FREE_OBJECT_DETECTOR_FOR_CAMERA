```python
# for idx in range(50, 65):
for idx in range(0, 10):
    
    img_path = selected_labels[idx]['img_path']
    class_labels = selected_labels[idx]['objCategory']
    bboxs = selected_labels[idx]['boundingBox2D']
    weather = selected_labels[idx]['weather']

    if len(class_labels) > 0:

        box_area = ( bboxs[:,2] - bboxs[:,0] ) * ( bboxs[:,3] - bboxs[:,1] )
        vizualize_remapped_annotations(
            img_path, bboxs, class_labels,
            figsize=(10, 5))
        
        print('weather   :', weather)
        for idx, label in enumerate(class_labels):
            print(f'{label} ......... box area {box_area[idx]}')
        
        print('=' * 100)

    else:
        print(f'No annotations presenst in image: {img_path}')
```
